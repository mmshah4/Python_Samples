## Using a web api
- Everything after the '?' is called the query string, '&' denotes variables
- the '{<variable_name>}' is special python string handling, allowing me to make a more descriptive string
### variables used
- prop: which information I want - in this case - the page extract
- format: format of the data, json / text / html / etc
- exsentences: How many sentences of the extract to return
- There is good documentation of this API in the links above

```python
def fetch_json_from_wiki(nom_url):
    #opener class reads the content of urls
    opener = urllib2.build_opener()
    #opener.addheaders = [('User-agent', 'Mozilla/5.0')]
    infile = opener.open(nom_url)
    return json.load(infile)

nom_text = fetch_json_from_wiki(wiki_base_extracts_url.format(num_sent=3, nom='Scott_Walker_(politician)'))
print nom_text
```

## Which candidate has the most republican wiki extract?

```python
from collections import Counter

def republican_index(extract):
    """
    function returns the gop term percentage
    """
    #Why am I replacing '.' and ','?, why lower case?
    words = extract.replace(".", "").replace(",", "").lower().split(" ")
    word_freq, total_words = Counter(words), len(words)
    republican_term_list = ['republican','gop','conservative']
    return sum([word_freq[x] for x in republican_term_list]) / float(total_words)
    
for nominee in nominee_extract:
    print nominee, "{:.2f}%".format(republican_index(nominee_extract[nominee]) * 100)
```
